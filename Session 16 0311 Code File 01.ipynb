### **Time Module**

```python
import time
```

- The function `time.time()` returns a float value representing the number of seconds that have elapsed since the epoch.
- The Epoch: The epoch is the fixed point in time where the time count begins.
- On most Windows and Unix systems, the epoch is defined as January 1, 1970, 00:00:00 Coordinated Universal Time (UTC).

```python
#----Current time of code execution---
time.time()
```

```python
def check_prime(x):
    for i in range(2,x):
        if x%i == 0:
            return f"{x} is Not Prime"
            break
    else:
        return f"{x} is Prime"
```

```python
check_prime(101)
```

```python
st = time.time()

check_prime(10000003)

et = time.time()
print(et-st)
```
